#### Groupby operation involves : SPLIT | APPLY FUCNTIONS | COMBINE RESULTS

```python
print(df['Country'].value_counts())
```

```python
country_grp = df.groupby(['Country'])  # splitting the data
```

```python
# df.loc[(df['Country'] =='India')]['SocialMedia'].value_counts()  # using filters
# df.loc[(df['Country'] =='India')]   # using filters
country_grp.get_group('India')        # using groups  

# country_grp
```

```python
country_grp['SocialMedia'].value_counts().head(50)
```

```python
country_grp['SocialMedia'].value_counts(normalize=True).loc['India']
```

```python
country_grp['ConvertedComp'].median().loc['India']
```

```python
country_grp['ConvertedComp'].agg(['median', 'mean']).loc['India']
```

```python
# using filter to find how many folks from India know python
filt= (df['Country'] =='India')
df.loc[filt]['LanguageWorkedWith'].str.contains('Python').sum()
```

```python
# country_grp['LanguageWorkedWith'].str.contains('Python').sum()  -> Error

country_grp['LanguageWorkedWith'].apply(lambda x: x.str.contains('Python').sum())
```

```python
country_uses_python = country_grp['LanguageWorkedWith'].apply(lambda x: x.str.contains('Python').sum())
country_respondent = df['Country'].value_counts()
country_respondent
```

```python
python_df = pd.concat([country_respondent, country_uses_python], axis='columns', sort=False)
python_df
```

```python
python_df.rename(columns={'Country': 'NumRespondents', 'LanguageWorkedWith': 'NumKnowsPython'}, inplace=True)
```

```python
python_df['PctKnowsPython'] = (python_df['NumKnowsPython']/python_df['NumRespondents'])*100
python_df
```

```python
python_df.sort_values(by='PctKnowsPython', ascending=False, inplace=True)
```
