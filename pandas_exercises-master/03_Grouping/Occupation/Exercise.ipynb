### Step 1. Import the necessary libraries

```python
import pandas as pd
```

### Step 2. Import the dataset from this [address](https://raw.githubusercontent.com/justmarkham/DAT8/master/data/u.user). 

### Step 3. Assign it to a variable called users.

```python
df = pd.read_csv('https://raw.githubusercontent.com/justmarkham/DAT8/master/data/u.user', sep='|')
df
```

### Step 4. Discover what is the mean age per occupation

```python
df.groupby('occupation').mean()['age']
```

### Step 5. Discover the Male ratio per occupation and sort it from the most to the least

```python
df.groupby(['occupation'], as_index=False).sum()
```

```python
df['m_count'] = df['gender'].apply(lambda x: 1 if x=='M' else 0)
df['f_count'] = df['gender'].apply(lambda x: 1 if x=='F' else 0)

df1 = df.groupby(['occupation', 'gender'], as_index=False)["m_count"].sum()
df1['count'] = df.groupby(['occupation', 'gender'], as_index=False)["m_count"].count()['m_count']
df1 = df1.groupby(['occupation']).sum()
df1['m_ratio'] = df1['m_count']/df1['count']*100
df1.sort_values('m_ratio', ascending=False)
```

### Step 6. For each occupation, calculate the minimum and maximum ages

```python
df.groupby('occupation')['age'].agg(['min', 'max'])
```

### Step 7. For each combination of occupation and gender, calculate the mean age

```python
df.groupby(['occupation', 'gender'])['age'].mean()
```

### Step 8.  For each occupation present the percentage of women and men

```python
# df.groupby(['occupation','gender'])['m_count', 'f_count'].sum()
men_df = df.groupby('occupation')['m_count'].sum() / df['occupation'].value_counts() * 100
women_df = df.groupby('occupation')['f_count'].sum() / df['occupation'].value_counts() * 100
percentage_df = men_df.to_frame(name='men_precentage')
percentage_df['women_precentage'] = women_df
percentage_df
```
